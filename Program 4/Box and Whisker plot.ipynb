# Box and Whisker Plot
    A box-and-whisker plot, also known as a box plot, is a standardized way of displaying the distribution of data based on a five-number summary: minimum, first quartile (Q1), median, third quartile (Q3), and maximum.

```python
import pandas as pd
import matplotlib.pyplot as plt

dataset = pd.read_csv('../ToyotaCorolla.csv')

plt.boxplot([dataset["Price"],dataset["HP"],dataset["KM"]])

plt.xticks([1,2,3],["Price","HP","KM"])
plt.title("Box and Whisker Plot")

plt.show()
```
