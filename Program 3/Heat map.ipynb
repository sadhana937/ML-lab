# Heat Map
    A heatmap is a graphical representation of data where individual values are represented by colors. It is useful for visualizing the magnitude of values across a matrix, typically to identify patterns, correlations, or anomalies in data.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

dataset = pd.read_csv("../ToyotaCorolla.csv")

sns.heatmap(dataset[["Price","KM","Doors", "Weight", "HP"]].corr(),cmap='jet')
plt.show()
```
